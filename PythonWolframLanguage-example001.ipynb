Import library and open a Wolfram Language session

```python
from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr

session = WolframLanguageSession()
```

Test some Wolfram Language code calls

```python
session.evaluate(wlexpr('N[1/5*Range[10]]'))
```

```python
session.evaluate(wlexpr('N[Pi,100]'))
```

```python
session.evaluate(wlexpr('Solve[x^2+3x+2==0,x]'))
```

```python
session.evaluate(wlexpr('D[x^2+3x+2,x]'))
```

```python
session.evaluate(wlexpr('Integrate[x^2+3x+2,x]'))
```

Make some Wolfram Alpha calls in plain English (NLP) to get some data-based answers

```python
session.evaluate(wlexpr('WolframAlpha["Who is Sonia Braga?", "SpokenResult"]'))
```

```python
session.evaluate(wlexpr('WolframAlpha["What is the meaning of life?", "SpokenResult"]'))
```

```python
session.evaluate(wlexpr('WolframAlpha["What is Caetano Veloso age?", "SpokenResult"]'))
```

Get images from Wolfram Language session

```python
image_data=session.evaluate(wlexpr('WolframAlpha["Christ the Redeemer", "Image",TimeConstraint->Infinity]'))
```

```python
type(image_data)
```

```python
picture=session.evaluate(wl.ExportByteArray(image_data, 'PNG'))
```

```python
type(picture)
```

```python
from PIL import Image
import io
```

```python
Image.open(io.BytesIO(picture))
```

Another example

```python
image_data=session.evaluate(wlexpr('WolframAlpha["NYC to Paris?", "Image",TimeConstraint->Infinity]'))
picture=session.evaluate(wl.ExportByteArray(image_data, 'PNG'))
Image.open(io.BytesIO(picture))
```

Close Wolfram session

```python
session.stop()
```
